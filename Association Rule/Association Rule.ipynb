```python
# importing libraries
from mlxtend.frequent_patterns import apriori,association_rules
```

### Reading and Understanding

```python
df = pd.read_csv('GroceryStoreDataSet.csv',header=None,names = ['Products'])
df.head()
```

```python
product = 'MILK,BREAD,BISCUIT'
product.split()
```

```python
# creating function 
def breakproductname(data):
    return data.split(',')

# calling the function
breakproductname(product)
```

```python
data = list(df['Products'].apply(breakproductname))
data
```

```python
# importing class for encoding
from mlxtend.preprocessing import TransactionEncoder

# creating instance
te = TransactionEncoder()

# encoding
enddata = te.fit_transform(data)
```

```python
print(te.columns_)
```

```python
data[0]
```

```python
print(enddata)
```

```python
df = pd.DataFrame(enddata,columns=te.columns_)
df.replace(True,1,inplace=True)
df.replace(False,0,inplace=True)
```

```python
df
```

### Model

```python
score = apriori(df,min_support=0.2,use_colnames=True)
score
```

```python
score_ar = association_rules(score,metric='confidence',min_threshold=0.5,num_itemsets='int')
score_ar
```
