```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = np.arange(-10,11)
x
```

```python
plt.figure(figsize=[6,3])
plt.plot(x)
plt.show()
```

```python
def sigmod(x):
    return 1 / (1 + np.exp(-x))
```

```python
y = sigmod(x)
```

```python
y
```

```python
plt.figure(figsize=[6,3])
plt.plot(x,y)
plt.show()
```
