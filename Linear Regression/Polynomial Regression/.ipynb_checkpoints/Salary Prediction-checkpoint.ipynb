### Import Required Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

#### Read and understand dataset

```python
# read data
df = pd.read_csv('Position_Salaries.csv')
df
```

```python
### plot
plt.scatter(df['Level'],df['Salary'])
plt.plot(df['Salary'])
```

```python
sns.regplot(data = df , x= df['Level'], y=df['Salary'])
```

```python
# split data into x and y
x = df[['Level']]
y = df['Salary']
```

#### Linear Regression Model

```python
# importing libraries
from sklearn.linear_model import LinearRegression

# creating instance
lr = LinearRegression()

# fit the model
lr.fit(x,y)

# predict 
ypred = lr.predict(x)
```

```python
# Evaluation metrics
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score

# r2 score
print(f'r2 score:{r2_score(y,ypred)}')

# mae
print(f'mean_absolute_error:{mean_absolute_error(y,ypred)}')

# mse
print(f'mean_squared_error:{mean_squared_error(y,ypred)}')

# rmse
print(f'root_mean_squared_error:{np.sqrt(mean_squared_error(y,ypred))}')
```
