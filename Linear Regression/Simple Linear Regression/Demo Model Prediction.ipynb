### Import Required Libraries

```python
import pandas as pd
import numpy as np
```

```python
dict1 = {
    'investment in lakhs':[1,2,3,4,5],
    'revenue in crore':[2,4,5,4,5]
}

df = pd.DataFrame(dict1)
```

```python
df
```

```python
df[['investment in lakhs']]
```

```python
df['revenue in crore']
```

```python
X = df[['investment in lakhs']] # 2d
y = df['revenue in crore'] # 1d
```

```python
# importing libraries
from sklearn.linear_model import LinearRegression

# create a instance/object
lr = LinearRegression()

# fitting our model
lr.fit(X,y)

# predicting 
ypred = lr.predict(X)
```

```python
ypred
```

```python
print(y,end=' ')
```

```python
# r2score
from sklearn.metrics import r2_score

r2_score(y,ypred)
```
