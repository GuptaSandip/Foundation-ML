### Importing Required Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Reading and understanding data

```python
df = pd.read_csv('data_titanic.csv')
df.head()
```

### dropping columns

```python
df.columns
```

```python
df.drop(columns = ['Unnamed: 0', 'PassengerId','Name','Ticket', 'Fare', 'Cabin','Initial'],inplace=True)
```

```python
df.head()
```

### Encoding

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

df['Sex'] = le.fit_transform(df['Sex'])
df['Embarked'] = le.fit_transform(df['Embarked'])
```

```python
df.head()
```

### Splitting data into x and y

```python
x = df.drop(columns = 'Survived')
y = df['Survived']
```

### Splitting data into train and test

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.2,random_state=1)
```

### Kernel
- Kernel : Kernel is used to handle non-linear data.
- kernel is also one of hyperparameter
- Type of kernel
- 1. Sigmoid - used in Layers(Neural network) and activation function.
  2. tanh - used in Layers(Neural network) and activation function.
  3. linear - when the data is linearly seperable we can use linear kernel.
  4. RBF(Radial basis kernel) - Gausian kernel mostly used when we don't have prior knowledge about the data.This is default kernel.
  5. Poly kernel - This kernel is widely used in image processing.

### Model1

```python
# import class
from sklearn.svm import SVC

# create instance 
svm = SVC()

# fit the model
svm.fit(xtrain,ytrain)

# predict
ypred = svm.predict(xtest)
```

```python
from sklearn.metrics import accuracy_score,classification_report
print(classification_report(ytest,ypred))
```

### Model2

```python
# import class
from sklearn.svm import SVC

# create instance 
svm = SVC(kernel='linear')

# fit the model
svm.fit(xtrain,ytrain)

# predict
ypred = svm.predict(xtest)
print(classification_report(ytest,ypred))
```

### Grid Search CV

- Helps in hypertunning making grid(table)
- It Automatically build model using hypertunning and in the end it will give us best model with hypertunning

```python
param_grid = {'C':[1,10,100],'gamma':[1,10,100]}
```

```python
from sklearn.model_selection import GridSearchCV

gs = GridSearchCV(svm,param_grid,verbose=3)

gs.fit(xtrain,ytrain)
```
